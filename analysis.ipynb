## experiment 4: 2, 3, 5

```python
print(rate4)
plt.scatter(df41.system_time, df41.new_clock_time - df41.old_clock_time, c='g', s=2, label="Rate = 2")
plt.scatter(df42.system_time, df42.new_clock_time - df42.old_clock_time, c='b', s=2, label="Rate = 3")
plt.scatter(df43.system_time, df43.new_clock_time - df43.old_clock_time, c='purple', s=2, label="Rate = 5")

plt.title("Experiment 4")
plt.xlabel("System Time (Nanoseconds)")
plt.ylabel("Logical Clock Jump Amount")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
print(rate4)

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(10, 3))

ax1.hist(df41.new_clock_time - df41.old_clock_time)
ax1.set_title("Rate = 2")

ax2.hist(df42.new_clock_time - df42.old_clock_time)
ax2.set_title("Rate = 3")

ax3.hist(df43.new_clock_time - df43.old_clock_time)
ax3.set_title("Rate = 5")

ax2.set_xlabel("Logical Clock Jump Amount Between Clock Cycles")
ax1.set_ylabel("Frequency")

# plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

# frequency of send vs receive

## experiment 7: 2, 4, 1

```python
labels = ['receive', 'send', 'internal']
colors = {'receive': 'gold',
           'send': 'tomato',
           'internal': 'skyblue'}
```

```python
def label_function71(val):
    return f'{val / 100 * len(df71):.0f}\n{val:.0f}%'

def label_function72(val):
    return f'{val / 100 * len(df72):.0f}\n{val:.0f}%'

def label_function73(val):
    return f'{val / 100 * len(df73):.0f}\n{val:.0f}%'


fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(10, 5))


df71['event_type'] = df71['event_type'].replace(['send1', 'send2', 'sendboth'], 'send')
df72['event_type'] = df72['event_type'].replace(['send1', 'send2', 'sendboth'], 'send')
df73['event_type'] = df73['event_type'].replace(['send1', 'send2', 'sendboth'], 'send')

df71.groupby('event_type').size().plot(kind='pie', autopct=label_function71, textprops={'fontsize': 10},
                                   ax=ax1,  title="Rate: 2", colors=[colors[key] for key in df71['event_type'].value_counts().keys()])

df72.groupby('event_type').size().plot(kind='pie', autopct=label_function72, textprops={'fontsize': 10},
                                  ax=ax2,   title="Rate: 4", colors=[colors[key] for key in df72['event_type'].value_counts().keys()])

df73.groupby('event_type').size().plot(kind='pie', autopct=label_function73, textprops={'fontsize': 10},
                                   ax=ax3,  title="Rate: 1",  colors=[colors[key] for key in df73['event_type'].value_counts().keys()])


```
