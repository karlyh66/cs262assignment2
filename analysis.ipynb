```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## experiment 4

```python
rate4 = [2, 3, 5]

df41 = pd.read_csv("data/4_machine1.csv")
df42 = pd.read_csv("data/4_machine2.csv")
df43 = pd.read_csv("data/4_machine3.csv")
```

```python
df4 = df41.copy(deep=True)
df4['machine'] = 1

df42['machine'] = 2
```

```python
plt.scatter(df43.iloc[:50].system_time, df43.iloc[:50].new_clock_time, c='g', s=2, label="Rate = 2")
plt.scatter(df42.iloc[:50].system_time, df42.iloc[:50].new_clock_time, c='b', s=2, label="Rate = 3")
plt.scatter(df41.iloc[:50].system_time, df41.iloc[:50].new_clock_time, c='purple', s=2, label="Rate = 5")

plt.title("Experiment 4")
plt.xlabel("System Time (Nanoseconds)")
plt.ylabel("Logical Clock Value")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
